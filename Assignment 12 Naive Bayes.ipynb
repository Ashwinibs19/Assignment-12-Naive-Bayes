# One Hot Encoding

```python
tr1 = df1.iloc[:,0:13]

tr1 = pd.get_dummies(tr1)
tr1
```

```python
te1 = test.iloc[:,0:13]

te1 = pd.get_dummies(te1)
te1
```

```python
#train data concat
finaltrain = pd.concat([tr1, df1['Salary']],axis=1)
finaltrain
```

```python
#test data concat
finaltest = pd.concat([te1, df2['Salary']],axis=1)
finaltest
```

# Dividing Finaltrain and Finaltest Data

```python
# Finaltrain data
X = finaltrain.values[:,0:102]
Y = finaltrain.values[:,102]

#Finaltest data
x = finaltest.values[:,0:102]
y = finaltest.values[:,102]
```

# Naive Bayes Model
Multinomial Naive Bayes

```python
classifier_mb = mb()
classifier_mb.fit(X,Y)
```

```python
y_mb_train = classifier_mb.predict(X)
```

```python
# training data
accuracy_score(Y,y_mb_train)
```

```python
confusion_matrix(Y,y_mb_train)
```

```python
print(classification_report(Y,y_mb_train))
```

```python
# testing data
y_test_mb = classifier_mb.predict(x)
```

```python
accuracy_score(y,y_test_mb)
```

```python
confusion_matrix(y,y_test_mb)
```

```python
print(classification_report(y,y_test_mb))
```

```python
print('training accurancy of Multinominal Naive Bayes model: ',accuracy_score(Y,y_mb_train))
print('testing accurancy of Multinominal Naive Bayes model: ',accuracy_score(y,y_test_mb))
```

# Gaussian Naive Bayes

```python
# training model
train_model = gb()
train_model.fit(X,Y)
```

```python
y_train_pred = train_model.predict(X)
```

```python
accuracy_score(Y,y_train_pred)
```

```python
confusion_matrix(Y,y_train_pred)
```

```python
print(classification_report(Y,y_train_pred))
```

```python
# testing model
y_test_pred = train_model.predict(x)
```

```python
accuracy_score(y,y_test_pred)
```

```python
confusion_matrix(y,y_test_pred)
```

```python
print(classification_report(y,y_test_pred))
```

```python
print('training data accurancy score is: ',accuracy_score(Y,y_train_pred))
print('testing data accurancy score is: ',accuracy_score(y,y_test_pred))
```
